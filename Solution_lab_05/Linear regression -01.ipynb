```python
print("By visualization by cma across the data is separated, So Median was best measure to fill the values ")
df['Trip_Price'] = df['Trip_Price'].fillna(df['Trip_Price'].median())
df['Trip_Distance_km'] = df['Trip_Distance_km'].fillna(df['Trip_Distance_km'].median())
```

```python

Q1 = df[['Trip_Distance_km', 'Trip_Price']].quantile(0.25)
Q3 = df[['Trip_Distance_km', 'Trip_Price']].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_no_outliers = df[~((df[['Trip_Distance_km', 'Trip_Price']] < lower_bound) | 
                      (df[['Trip_Distance_km', 'Trip_Price']] > upper_bound)).any(axis=1)]
```
